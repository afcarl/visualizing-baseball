```python
%matplotlib inline
import pybaseball as pb
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
pitching_data = pb.team_pitching(1900, 2015)

pitching_data.head()
```

```python
pitching_data = pitching_data[['Season', 'W', 'L', 'K/9', 'BB/9']]

pitching_data['GP'] = pitching_data['W'] + pitching_data['L']

pitching_data.head()
```

```python
seasons = pitching_data['Season'].tolist()
games = pitching_data['GP'].tolist()
k9 = pitching_data['K/9'].tolist()
bb9 = pitching_data['BB/9'].tolist()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(seasons, k9)
plt.xlabel('Season')
plt.ylabel('Strike Outs 9 Innings per Team per Game ')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(seasons, bb9)
plt.xlabel('Season')
plt.ylabel('Walks per 9 Innings per Team per Game')
plt.show()
```

```python
pitching_data.groupby(['Season']).mean().head(5)
```

```python
ak9ptps = pitching_data.groupby(['Season']).mean()['K/9'].tolist() #average runs per team per season
abb9ptps = pitching_data.groupby(['Season']).mean()['BB/9'].tolist() #average games per team per season
s = list(range(1900, 2016))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(seasons, k9)
plt.plot(s, ak9ptps, 'b', linewidth=3)
plt.xlabel('Season')
plt.ylabel('Strike Outs 9 Innings per Team per Game ')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(seasons, bb9)
plt.plot(s, abb9ptps, 'b', linewidth=3)
plt.xlabel('Season')
plt.ylabel('Walks per 9 Innings per Team per Game')
plt.show()
```
